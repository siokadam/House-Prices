<font size = 2><b>2.1.3. Missing values imputation</b></font>

To impute missing values it is needed to understand data which is explained in data_descirption.txt file.<br>

There are 2 types of missing data in categorical columns:<br>
Part of the house - 'PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageFinish', 'GarageQual', 'GarageCond', 'GarageType', 'BsmtExposure', 'BsmtFinType2', 'BsmtCond', 'BsmtQual', 'BsmtFinType1', 'MasVnrType'<br>
Media - 'Electrical', 'MSZoning', 'Utilities', 'Functional', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'SaleType'<br>
<br>
NaN value in first type probably means that such part doesn't exist in house. These values will be imputed with <b><i>None</i></b><br>
NaN value in second type is connected with media. Every house has a media so here a <b><i>mode</i></b> value will be imputed.<br><br>
There are 2 types of missing data in numerical columns:<br>
Linear feet of street connected to property - 'LotFrontage'<br>
Features that is connected with their home parts existence - 'GarageYrBlt', 'MasVnrArea', 'BsmtFullBath', 'BsmtHalfBath', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'GarageCars', 'GarageArea'<br><br>
NaN value in first type is a true missing data. It will be imputed with <b>mean</b>.<br>
NaN value in second type is connected with a part of the house. If it is then mean value will be imputed, else 0.

```python
def impute_nans(df, cols, how='NA', col_to_check=None):
    """Function to impute missing values in a DataFrame.
    
    Args:
        df (DataFrame): DataFrame to impute missing values.
        cols (list): Columns to impute missing values.
        how (string): How to impute missing values.
        col_to_check (string): Column which should be checked before imputing 0 or mean value,
                                e.g. for basement square column it checks if basement exists. If yes then impute mean, else 0.
        
    Returns:
        DataFrame.
    """
    if how == 'NA':
        for col in cols:
            df[col].fillna('NA', inplace=True)
    elif how == 'mode':
        for col in cols:
            df[col].fillna(df[col].mode()[0], inplace=True)
    elif how == 'mean':
        for col in cols:
            df[col].fillna(df[col].mean(), inplace=True)
    elif how == 'check':
        for col in cols:
            df.loc[(df[col_to_check] != 'NA') & ~df[col_to_check].isna() & (df[col].isna()), col] = df[col].mean()
            df.loc[((df[col_to_check] == 'NA') | df[col_to_check].isna()) & (df[col].isna()), col] = 0
```

```python
cat_type_1 = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageFinish', 'GarageQual', 'GarageCond', 'GarageType', 'BsmtExposure', 'BsmtFinType2', 'BsmtCond', 'BsmtQual', 'BsmtFinType1', 'MasVnrType']
cat_type_2 = ['Electrical', 'MSZoning', 'Utilities', 'Functional', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'SaleType']
num_type_1 = ['LotFrontage']
num_type_2_1 = ['MasVnrArea']
num_type_2_2 = ['BsmtFullBath', 'BsmtHalfBath', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']
num_type_2_3 = ['GarageCars', 'GarageArea', 'GarageYrBlt']

for df in [df_train, df_test]:
    impute_nans(df, cat_type_1, 'NA')
    impute_nans(df, cat_type_2, 'mode')
    impute_nans(df, num_type_1, 'mean')
    impute_nans(df, num_type_2_1, 'check', 'MasVnrType')
    impute_nans(df, num_type_2_2, 'check', 'BsmtCond')
    impute_nans(df, num_type_2_3, 'check', 'GarageCond')
```

<font size = 3><b>2.3. Numeric variables manipulation</b></font>

<font size = 2><b>2.3.1. Features creation</b></font>

As there are some numeric values that are related to each other, let's create some new common columns.

```python
for df in [df_train, df_test]:
    df['AvgQuality'] = (df['OverallQual'] + df['OverallCond']) / 2
    df['IsRemodeled'] = (df['YearRemodAdd'] - df['YearBuilt']).apply(lambda x: 1 if x > 0 else 0)
    df['Total1st2ndSF'] = df['1stFlrSF'] + df['2ndFlrSF']
    df['TotalFullBath'] = df['BsmtFullBath'] + df['FullBath']
    df['TotalHalfBath'] = df['BsmtHalfBath'] + df['HalfBath']
    df['GarageAreaPerCar'] = (df['GarageArea'] / df['GarageCars']).fillna(0)
    df['TotalPorch'] = df['OpenPorchSF'] + df['EnclosedPorch'] + df['3SsnPorch'] + df['ScreenPorch']
    df['SoldHouseAge'] = df['YrSold'] - df['YearBuilt']
    df['GarageAge'] = (df['YrSold'] - df['GarageYrBlt']).apply(lambda x: 0 if x > 1500 else x)
```

<font size = 2><b>2.3.2. Features deletion</b></font>

```python
for df in [df_train, df_test]:
    df.drop(['OverallQual', 'OverallCond', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath',
       'HalfBath', 'GarageCars', 'GarageArea', 'GarageYrBlt', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch',
       'YrSold', 'YearBuilt'], axis=1, inplace=True)
```

<font size = 5><b>3. Model building</b></font>

<font size = 3><b>3.1. Splitting dataset</b></font>

```python
X_training = df_train.drop(['Id', 'SalePrice'], axis=1)
y_training = df_train['SalePrice']

X_testing = df_test.drop('Id', axis=1)
```

<font size = 3><b>3.2. Scaling data</b></font>

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_training)
X_test_scaled = scaler.transform(X_testing)

y_training_min = y_training.min()
y_training_max = y_training.max()

y_train_scaled = (y_training - y_training.min()) / (y_training.max() - y_training.min())

X_train, X_test, y_train, y_test = train_test_split(X_train_scaled, y_train_scaled, test_size=0.3, random_state=1712)
```

<font size = 3><b>3.3. Instantiate models</b></font>

```python
kf = KFold(n_splits=5, shuffle=True, random_state=1712)
linreg = LinearRegression()
lasso = Lasso()
ridge = Ridge()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor(random_state=1712)
gbr = GradientBoostingRegressor(random_state=1712)
models = {'LinearRegression':linreg, 'Lasso':lasso, 'Ridge':ridge,'DecisionTreeRegressor':dtr, 'RandomForestRegressor':rfr, 'GradientBoostingRegressor':gbr}
```

<font size = 3><b>3.4. Cross validation</b></font>

```python
for name, model in models.items():
    cv = - cross_val_score(model, X_train, y_train, cv=kf, scoring='neg_root_mean_squared_error')
    print(f'{name} RMSE: {cv.mean()}')
```

GradientBoostingRegressor shows the best performance beacuse it has the lowest RMSE value. Let's hyperparameter it with GridSearchCV.

<font size = 3><b>3.5. Hyperparameter optimization</b></font>

```python
params = {'learning_rate':[0.01, 0.1],
          'n_estimators':[100, 200, 400],
          'max_depth':[2, 3]}
gs = GridSearchCV(gbr, param_grid=params, cv=kf)
gs.fit(X_train, y_train)
print('Best score: ', gs.best_score_)
print('Best hyperparameters: ', gs.best_params_)
```

<font size = 3><b>3.6. Model results</b></font>

```python
y_pred = gs.predict(X_test)
print('Train RMSE: ', MSE(y_test, y_pred, squared=False))
```

<font size = 5><b>4. Submission</b></font>

<font size = 3><b>4.1. Predicting data</b></font>

```python
y_pred_scal = gs.predict(X_test_scaled)
y_prediction = y_pred_scal * (y_training_max - y_training_min) + y_training_min
```

<font size = 3><b>4.2. Exporting data</b></font>

```python
submission = pd.DataFrame({'Id':df_test['Id'], 'SalePrice':y_prediction})
submission.to_csv('house_prices_submission.csv', index=False)
```
